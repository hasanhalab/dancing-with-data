# Feature Extraction

```python
def estimate_seating_capacity(row):
    '''
    
    This function estimates the total seating capacity based on the relevant features
    ['doors', 'persons', 'Cost']
    
    
    '''
    
    # Extract relevant features
    doors = row["doors"]
    persons = row["persons"]
    cost = row["Cost"]

    seating_capacity = None

    # Apply logic based on features
    if doors == '2':
        if persons == '2':
            seating_capacity = '2'
        else:
            seating_capacity = '4'
    elif doors == '3':
        seating_capacity = '4'
    elif doors >= '4':
        if cost == "low" or "med":
            seating_capacity = '5'
        else:
            seating_capacity = '6'
    
    return seating_capacity

# Apply the function to each row and add a new "total_seating_number" column
df["total_seating_number"] = df.apply(estimate_seating_capacity, axis=1)
```

```python
df
```

```python
def is_family_friendly(row):
    '''
    
    This function determines whether the car is family friendly or not based on these features
    ['total_seating_number', 'luggage_boot', 'safety']
    
    
    '''
    # Extract relevant features
    total_seating_number = row["total_seating_number"]
    cargo_space = row["luggage_boot"]
    safety_rating = row["safety"]

    # Define minimum thresholds
    min_seating_capacity = '5'
    min_cargo_space = "big"
    min_safety_rating = "high" or 'med'

    # Apply criteria and assign weights
    family_friendliness = 0
    if int(total_seating_number) >= int(min_seating_capacity):
        family_friendliness += 2
    if cargo_space == min_cargo_space:
        family_friendliness += 1
    if safety_rating == min_safety_rating:
        family_friendliness += 2

    # Set a threshold for family-friendliness score
    
    family_friendliness_threshold = 3

    # Determine family-friendly status based on the score
    if (family_friendliness >= family_friendliness_threshold) and (int(total_seating_number) >= int(min_seating_capacity)):
        return "Yes"
    else:
        return "No"

# Apply the function to each row and add a new "family_friendly" column
df["family_friendly"] = df.apply(is_family_friendly, axis=1)
```

```python
# split X and y into training and test sets
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.33, random_state=42)
```

# Feature Engineering

```python
X_train.head()
```

```python
# encoding process in categorical columns
#!pip install category_encoders
import category_encoders as ce
```

```python
# encode categorical variables with ordinal encoding

encoder = ce.OrdinalEncoder(cols=['Cost', 'maintenance', 'doors', 'persons', 'luggage_boot', 'safety', 'total_seating_number'
                                 , 'family_friendly'])

X_train = encoder.fit_transform(X_train)

X_test = encoder.transform(X_test)
```

```python
X_train.head()
```

```python
X_test.head()
```

```python
'''
StandardScaler is a technique used in machine learning 
and statistics for feature scaling or normalization. 
Feature scaling is a crucial preprocessing step that 
helps in improving the performance and stability of 
many machine learning algorithms. 

'''
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

#                      (20 hidden layers, 12 neurons)
mlp = MLPClassifier(hidden_layer_sizes=(20, 12), max_iter=150, random_state=42, learning_rate_init = 0.002)
mlp.fit(X_train_scaled, y_train)

'''
MLPClassifier  uses (tanh) as the default activation function 
for hidden layers and the (softmax) function for the output layer for multiclass classification problems.

'''

y_pred_train = mlp.predict(X_train_scaled)
y_pred_test = mlp.predict(X_test_scaled)

accuracy_training = accuracy_score(y_train, y_pred_train)
accuracy_test = accuracy_score(y_test, y_pred_test)

conf_matrix = confusion_matrix(y_test, y_pred_test)

print("Accuracy (Training data): {:.2f} %".format(accuracy_training*100))
print("Accuracy (Testing data): {:.2f} %".format(accuracy_test*100))
print("Confusion Matrix:")
print(conf_matrix)


plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=mlp.classes_, yticklabels=mlp.classes_)
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix (MLP)')
plt.show()
```

```python
training_loss = mlp.loss_curve_
'''
the loss_curve_ attribute provides information about 
the training loss at each iteration (epoch) during the training process.

'''
# Plotting the loss for each epoch
plt.figure(figsize=(8, 4))
plt.plot(training_loss, label='Training Loss', color='blue')
plt.title('Training Loss Over Epochs')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
# Plotting the average loss by the number of epochs

average_loss_by_epoch = np.cumsum(training_loss) / np.arange(1, len(training_loss) + 1)

plt.figure(figsize=(8,4))
plt.plot(average_loss_by_epoch, label='Average Training Loss', color='red')
plt.title('Average Training Loss Over Epochs')
plt.xlabel('Epochs')
plt.ylabel('Average Loss')
plt.legend()
plt.show()
```

# Feature Engineering

```python
# why we using the SMA in the time series prediction??

# In time series prediction, Simple Moving Average (SMA) often plays a crucial role in:

# Smoothing out noise and highlighting underlying trends: By averaging recent price values, SMA filters out short-term
# fluctuations and reveals the broader pattern of price movement. This clearer view of the trend can improve model accuracy
# and reduce sensitivity to random noise.

# Serving as a benchmark for comparison: In the given code, the 100-day SMA acts as a reference line against which to
# compare the model's predictions. This helps assess whether the model is accurately capturing the underlying trend and
# identifies potential deviations.

# Capturing momentum and potential reversals: Changes in the direction of the SMA can signal shifts in market momentum or
# potential trend reversals. Incorporating this information can enhance the model's ability to anticipate future price
# movements.

# Generating trading signals: SMAs are often used directly to generate buy and sell signals in trading strategies.
# For example, a crossover of the price above the SMA might suggest a buy signal, while a crossover below might indicate
# a sell signal.

sma_window = 100    # SMA (Simple Moving Average)
df['SMA_100'] = df['Close'].rolling(window=sma_window).mean()
```

```python
df['SMA_100'].head(99)
```

```python
# why there are 100 of nulls values ??

# The rolling window of 100 days is specified for calculating the SMA. As a result, for the first 99 days of the dataset,
# there are not enough preceding days to compute the average, leading to null values in the 'SMA_100' column for these
# initial data points. This is a common occurrence when applying rolling window calculations, and it's necessary to
# handle or exclude these null values appropriately based on the specific requirements of your analysis.

df.isna().sum()
```

```python
# why we using the EMA in the time series prediction??

# Exponential Moving Average (EMA) is used in time series prediction for its responsiveness to recent changes in data
# compared to Simple Moving Average (SMA). The provided code calculates the EMA for a given window size (ema_window).
# EMA gives more weight to recent observations, making it suitable for capturing trends and reacting quickly to changes in
# the underlying data. This can be beneficial in time series prediction tasks where recent trends or patterns are essential
# for making accurate predictions.

ema_window = 36

# The smoothing factor (alpha) determines the weight given to the most recent data points in the calculation
alpha = 2 / (ema_window + 1) 
df['EMA_36'] = df['Close'].ewm(alpha=alpha, adjust=False).mean()
```

```python
df.isna().sum()
```

### What is the difference between EMA and SMA?

#### In time series analysis, the choice between Simple Moving Average (SMA) and Exponential Moving Average (EMA) hinges on their distinct weighting approaches. SMA treats all data points equally within a specified window, resulting in a smoother average but potentially slower responsiveness to recent changes. On the contrary, EMA assigns exponentially decreasing weights to data points, giving more significance to recent observations and making it more responsive to short-term fluctuations. The formulas for SMA involve a straightforward average calculation, while EMA employs a smoothing factor for weighted averaging. Essentially, SMA provides stability, while EMA captures short-term trends with greater sensitivity to recent data, allowing for flexibility based on the analytical needs of the time series data.

### Is the window size  important for the EMA and SMA?

#### Yes, the window size is crucial for both Simple Moving Average (SMA) and Exponential Moving Average (EMA) in time series analysis. The window size determines the number of data points considered when calculating the moving average. For SMA, the window size directly corresponds to the number of periods over which the average is computed. Larger window sizes result in smoother averages but might lag behind significant changes. In contrast, EMA's responsiveness is controlled by its smoothing factor, which indirectly depends on the window size through the formula involving the exponential decay. Smaller window sizes in EMA give more weight to recent observations, making it more responsive to short-term trends. Therefore, selecting an appropriate window size is a critical decision that balances the need for smoothness and responsiveness based on the characteristics of the time series data.

```python
# why we using the RSI in the time series prediction??

# The Relative Strength Index (RSI) is employed in time series prediction, as indicated in the provided code snippet,
# to gauge the momentum and potential overbought or oversold conditions in a financial instrument's price. RSI is calculated
# based on the average gain and loss over a specified window, commonly 14 periods in this case. By comparing the magnitude
# of recent gains and losses, RSI generates values that oscillate between 0 and 100. Low RSI levels, typically below 30,
# suggest potential buying opportunities, signaling that the asset may be oversold. Conversely, high RSI levels, typically
# above 70, indicate potential selling opportunities, suggesting that the asset may be overbought. In summary, RSI serves
# as a valuable tool for identifying potential trend reversals and guiding trading decisions based on short-term market
# conditions.


rsi_window = 14
price_diff = df['Close'].diff(1)
gain = price_diff.where(price_diff > 0, 0)
loss = -price_diff.where(price_diff < 0, 0)

avg_gain = gain.rolling(window=rsi_window, min_periods=1).mean()
avg_loss = loss.rolling(window=rsi_window, min_periods=1).mean()

rs = avg_gain / avg_loss
rsi = 100 - (100 / (1 + rs))

df['RSI_14'] = rsi
```

```python
# In this code, we are preparing the input sequences (X) and corresponding output values (Y) for a time series prediction
# task. The variable time_steps is set to 30, indicating that each input sequence in X will consist of 30 time steps.
# We iterate through the time series data (y) starting from the 31st data point (since we need 30 previous points for each
# sequence) up to the end. For each iteration, we extract a sequence of 30 consecutive data points as input (X) and the next
# data point as the output (Y). The resulting X and Y are converted to NumPy arrays, and the shape of X is adjusted to be
# compatible with LSTM input requirements. The printed output indicates the final shape of X and Y, which are crucial for
# building and training LSTM models.


# for the first 30 values of X, the value number 31 is append to y,
# for the second  30 values of X, the value number 32 is append to y,
# for the third 30 values of X, the value number 33 is append to y,
# for the fourth 30 values of X, the value number 34 is append to y,
# for the fifth 30 values of X, the value number 35 is append to y,
# for the sixth 30 values of X, the value number 36 is append to y and so on.

time_steps = 30
X, Y = [], []                       # len(time_steps) /len(y)
for i in range(time_steps, len(y)): # from 30 to 5703
    X.append(y[i - time_steps:i, 0])
    Y.append(y[i, 0])
X, Y = np.array(X), np.array(Y)
X = np.reshape(X, (X.shape[0], X.shape[1], 1))
print("X.shape", X.shape, "Y.shape", Y.shape)
```

```python
from sklearn.linear_model import LinearRegression


data = y.reshape(-1, 1)

input_size = 1
reservoir_size = 100
output_size = 1
leak_rate = 0.3
spectral_radius = 0.9

train_ratio = 0.8
train_size = int(train_ratio * len(data))

train_data, test_data = data[:train_size], data[train_size:]

input_weights = np.random.rand(reservoir_size, input_size) - 0.5
reservoir_weights = np.random.rand(reservoir_size, reservoir_size) - 0.5
reservoir_weights *= spectral_radius / max(np.abs(np.linalg.eigvals(reservoir_weights)))
output_weights = np.random.rand(output_size, reservoir_size) - 0.5

reservoir_states = np.zeros((reservoir_size, 1))

output = np.zeros_like(data)

lagged_features = np.zeros((train_size, reservoir_size + 1))
lagged_features[:, 0] = 1  # Bias term
lagged_features[:, 1:] = reservoir_states[:train_size].T

# Run the ESN for training
for t in range(train_size):
    input_state = train_data[t, :].reshape((input_size, 1))
    reservoir_states = (1 - leak_rate) * reservoir_states + leak_rate * np.tanh(
        np.dot(reservoir_weights, reservoir_states) + np.dot(input_weights, input_state)
    )
    lagged_features[t, 1:] = reservoir_states.T

# Use Linear Regression for output weights training
lr = LinearRegression()
lr.fit(lagged_features, train_data.flatten())
output_weights_train = lr.coef_[1:].reshape(output_size, reservoir_size)

# Run the ESN for testing
for t in range(train_size, len(data)):
    input_state = data[t, :].reshape((input_size, 1))
    reservoir_states = (1 - leak_rate) * reservoir_states + leak_rate * np.tanh(
        np.dot(reservoir_weights, reservoir_states) + np.dot(input_weights, input_state)
    )
    lagged_feature_test = np.concatenate([reservoir_states.flatten()])
    output[t, :] = np.dot(output_weights_train, lagged_feature_test)

mse_train = mean_squared_error(train_data, output[:train_size])
print(f"Mean Squared Error on Training Data: {mse_train}")    
    
    
# Evaluate the model performance using MSE
mse = mean_squared_error(data[train_size:], output[train_size:])
print(f"Mean Squared Error on Test Data: {mse}")

# Plot only the test data along with predicted values
plt.figure(figsize=(18,6), facecolor='black')
actual_prices = scaler_y.inverse_transform(data[train_size:])
predicted_prices = scaler_y.inverse_transform(output[train_size:-1])

plt.plot(df.index[train_size:train_size+len(actual_prices)], actual_prices, label="Actual Close Prices", color="red")
plt.plot(df.index[train_size:train_size+len(predicted_prices)], predicted_prices, label="Predicted Close Prices", color="gold", linestyle='dashed')

plt.title("Actual vs. Predicted (ESN)", fontsize=18, color='gold', fontweight='bold')
plt.xlabel("Timestamp", fontsize=14, color='gold')
plt.ylabel("Closing Price", fontsize=14, color='gold')
plt.xticks(fontsize=12, rotation=45, color='white')  

plt.yticks(fontsize=12, color='white')
plt.grid(True, linestyle='--', alpha=1, color='gray')
plt.box(False)

plt.legend()
plt.show()



future_steps = 90     # prediction for next 3 months

future_predictions = np.zeros((future_steps, 1))

reservoir_states_future = reservoir_states[:, -1].reshape((reservoir_size, 1))

# Run the ESN for future predictions
for t in range(future_steps):
    input_state_future = np.random.rand(input_size, 1)  

    reservoir_states_future = (1 - leak_rate) * reservoir_states_future + leak_rate * np.tanh(
        np.dot(reservoir_weights, reservoir_states_future) + np.dot(input_weights, input_state_future)
    )

    lagged_feature_future = np.concatenate([reservoir_states_future.flatten()])

    future_predictions[t, :] = np.dot(output_weights_train, lagged_feature_future)

future_dates = pd.date_range(start='2022-09-03', periods=future_steps, freq='D')

# Plot only the future predictions plot
plt.figure(figsize=(18, 6), facecolor='black')  
plt.plot(future_dates, scaler_y.inverse_transform(future_predictions), label="Future Predictions", color="red", linestyle='dashed')
plt.xlabel('Date', color='gold',fontsize=14)
plt.ylabel('Closing Price', color='gold',fontsize=14)
plt.legend()
plt.title('Future Predictions (3 months)', color='gold', fontsize=18, fontweight='bold')
plt.xticks(rotation=45, color='white',fontsize=12)
plt.yticks(color='white',fontsize=12)
plt.grid(True, linestyle='--', alpha=0.4, color='gray')

plt.box(False)
plt.xlim([pd.Timestamp('2022-09-01'), pd.Timestamp('2022-12-01')])

plt.show()
```
